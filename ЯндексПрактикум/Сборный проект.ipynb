```python
#отберем n-ое колличево наиболее продаваемых платформ
def best_platform (df, number): 
    return (
        df
        .pivot_table(index='platform', values='combined_sales', aggfunc='sum')
        .sort_values('combined_sales', ascending=False)
        .head(number)
        .index.get_level_values(0)
    )  
number_of_best_platform = 6
best_platform = best_platform(games,number_of_best_platform)
```

```python
# построим динамику продаж наиболее продаваемых платформ
(games[games['year_of_release'] != 2100]
     .query('platform in  @best_platform')
     .pivot_table(index='year_of_release', values='combined_sales', columns = 'platform', aggfunc='sum')
     .plot(kind='line', figsize=(10, 5), fontsize=10)
     .set(xlabel='Год', ylabel='Суммарные продажи', title ="Динамика продаж игр" ))
plt.xlim([1999, 2019])
```

```python
# определим фунция жизненного цикла платформ
def lifecycle_of_games (df,platform):
    cols = ['Игра', 'Год начала выпуска','Год окончания выпуска', 'Кол-во лет в продаже' ]
    new_table = []
    for i in range(len(platform)):
        y1 = platform[i]
        table = df[df['year_of_release'] != 2100].query('platform == @y1')
        
        y2 = table['year_of_release'].max() 
        y3 = table['year_of_release'].min()
        y4 =  table['year_of_release'].max() - table['year_of_release'].min()
        new_table.append([y1,y3, y2,y4])
    return  pd.DataFrame(new_table, columns=cols)
```

```python
(lifecycle_of_games(games,games['platform'].unique())
.sort_values(by = "Кол-во лет в продаже", ascending =False)
.head(10)
)
```

<span style="color:red">Вывод: Популярность платформ циклична и в среднем равна 10 годам. Так, например, PS2 была некогда популярной платформой, однако к 2012 г. продажи сократились до нуля. Одна есть и выбросы равные 31 и 28 годам популярности.

```python
# отбор данных за период с 2012 по 2106
researched_period = (games['year_of_release'] >= 2012) & (games['year_of_release'] <= 2016)
researched_games = games.loc[researched_period].reset_index(drop=True)
```

```python
#отберем 6 платформ в обновленном датафрейме researched_games
number = 6 
best_researched = (
        researched_games
        .pivot_table(index='platform', values='combined_sales', aggfunc='sum')
        .sort_values('combined_sales', ascending=False)
        .head(number)
        .index.get_level_values(0)
                  )
```

```python
# построим графики измения суммарных продаж
rows = 3
cols = (len(best_researched) // rows)
fig, ax = plt.subplots(rows, cols)

# построение графиков продаж по каждой платформе
num = 0
for i in range(rows):
    for j in range(cols):
        y = best_researched[num]
        (researched_games
         .query('platform in @y')
         .pivot_table(index='year_of_release', values='combined_sales', columns = 'platform', aggfunc='sum')
         .plot(ax=ax[i, j], figsize=(20, 20) )
            )
        num += 1 
plt.tight_layout()
plt.show()
```

## Шаг 4.  Портрет пользователя каждого региона

```python
#  рассмотрим наиболее популярные платформы по регионам
def top5_in_region(df, column, region):
    df = (df
            .pivot_table(index=column, values=region, aggfunc='sum')
            .sort_values(region, ascending=False)
            .head(5)
         )
    
    
    (df
     .plot(kind='bar', y=region, figsize=(10, 5), legend=False)
     .set(ylabel='Продажи ' + region)
      
    )
    plt.title('Самые популярные платформы (топ-5)')
```

```python
top5_in_region(researched_games,column = researched_games['platform'],region = 'jp_sales')
top5_in_region(researched_games,column = researched_games['platform'],region = 'eu_sales')
top5_in_region(researched_games,column = researched_games['platform'],region = 'na_sales')
```

## Шаг 6. Проверка гипотез

1) Нулевая гипотеза H₀: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Исходя из H₀ формулируем альтернативную гипотезу H₁: Средние пользовательские рейтинги платформ Xbox One и PC разные.

```python
#define F-test function
def f_test(x, y):
    x = np.array(x)
    y = np.array(y)
    f = np.var(x, ddof=1)/np.var(y, ddof=1) #calculate F test statistic 
    dfn = x.size-1 #define degrees of freedom numerator 
    dfd = y.size-1 #define degrees of freedom denominator 
    p = 1-st.f.cdf(f, dfn, dfd) #find p-value of F test statistic 
    return f, p
```

```python
alpha = 0.05
xone = researched_games.query("platform == 'XOne'").dropna()
pc = researched_games.query("platform == 'PC'").dropna()
x1, y1 = xone['user_score'], pc['user_score']
p = f_test(x1, y1)[1]

if p < alpha:
    print("Отвергаем нулевую гипотезу о равенстве дисперсий")
else:
    print("Не получилось отвергнуть нулевую гипотезу  о равенстве дисперсий") 
```

```python
alpha = 0.1 # так  как 50% пропуски

results = st.ttest_ind(
    x1, y1,
    equal_var= True, 
    nan_policy='omit' # игнорируем пропуски
)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу о равенстве средних")
else:
    print("Не получилось отвергнуть нулевую гипотезу о равенстве средних")
```
