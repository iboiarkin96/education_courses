# Проведение A/B testing

```python
from IPython.display import Image
```

**Схема А/В тестирования**
1. Берем по 1% пользователей 
2. Разбираем этих пользователей случайно на 2 группы 
	- одной группе накатываем изменения 
	- на другой оставляем как есть 
3. Считаем метрики и оцениваем их среднее значение $\overline{x_b}$ и $\overline{x_а}$
4. Вычисляем кретирий для оценки $\Delta {X} =\overline{x_b} - \overline{x_а}$
	- *существенность* $\Delta {X}$ vs. 0 ->  позитивные или негативные изменения (так как новый алгоритм может быть дорогим, а сущесвенность его может быть мала)
	- *значимость*: статистический тест -> разница вызвана шумом (плохо) и измененими(хорошо)
5. Можно новую версию показать 95% пользователей, а старую - 5% - **обратное А/В тестирования**

**Значимость** - статистический тест говорит нам, о том, что изменения не случайны 

**Существенность** - насколько измениния **больше** по своей величине, насколько размер эффект(измкенкени яметрики), который мы ловим, 

**Типичные пользовательские мерики**
 - клики
 - срднее время пользователя 
 - возвращаемость 
 - средний чек
 - средний трафик
 

**Где используется?**
 - изменение дизайна 
 - изменение функциональности в играх 
 - рабостоспособность лекарст (настоящее против плацебо)
 - новый алгоритм машинного обучения
 - изменения в офлайн- магазинах: смена порядка отделов, раскладка товаров, установка постоматов, промоауции
