```python
# вывод элементов чисел фибаначи 
def fib(n):
	result =[0, 1]
	if n == 1:
		return result[0]
	elif n == 2:
		return result
	else:
		for i in range(2, n):
			a = result[i-2] + result[i-1]
			result.append(a)
		return result
```

- inheritance (наследование) -это процесс, когда один класс наследует атрибуты и методы другого. (срш)
- Инкапсуляция - связь данных с методами которые этими данными управляют (например, машина, мы видим окна, но не видим двигатель)
- Абстракция – это выделение основных, наиболее значимых характеристик объекта и игнорирование второстепенных. 
- Полиморфизм - это способность выполнять действие над объектом независимо от его типа.


```python
class Dog:
	def __init__(self, name, age) -> None:
		self.name = name
		self.age = age
		self.food = 'meet'
	
	def sit(self):
		print(self.name.title() + ' is sitting')
# инициализация класса
s = Dog(name = 'Пес', age=15)		
s.sit()
```

```python
# наследование
class Dog_Parent: # родительский класс

	sex =  'man'
	def __init__(self, ) -> None:
		print('init parent')
	
	def parent_met(self):
		print('par metod')



class Dog_Child(Dog_Parent): # child class

	'''
	так как инин есть в родители и в child то в child init перезапишет parent
	'''
	def __init__(self, ) -> None:
		print('init child')
	
	def chil_met(self):
		print('chil metod')

Dog_Child().sex
```

```python
# абстракция 
from abc import ABC, abstractmethod
class Animal(ABC):
	
	@abstractmethod
	def eats_grass(self):
		pass
	
	def is_animal(self):
		return True

class Cat(Animal):
	def __init__(self) -> None:
		print('мяу')
	# запись абстрактного сметода 
	def eats_grass(self):
		return False

m = Cat()
print(m.is_animal())
print(m.eats_grass())
```

```python
# полиморфизм

class Ivan():
	def language(self):
		print('python')


class Vova():
	def language(self):
		print('all')	

i = Ivan()
v = Vova()	
for j in [i, v]:
	j.language()
```
