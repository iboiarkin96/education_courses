```python
# вывод элементов чисел фибаначи 
def fib(n):
	result =[0, 1]
	if n == 1:
		return result[0]
	elif n == 2:
		return result
	else:
		for i in range(2, n):
			a = result[i-2] + result[i-1]
			result.append(a)
		return result
```

```python
# абстракция 
from abc import ABC, abstractmethod
class Animal(ABC):
	
	@abstractmethod
	def eats_grass(self):
		pass
	
	def is_animal(self):
		return True

class Cat(Animal):
	def __init__(self) -> None:
		print('мяу')
	# запись абстрактного сметода 
	def eats_grass(self):
		return False

m = Cat()
print(m.is_animal())
print(m.eats_grass())
```

```python
def y(x):
    return 0.5*x**3 - 5*x**2 + 0.5*x - 5
```

```python
a = 6
b = 7
epsilon = 10 ** (-1)
iterations = 1
while np.abs(a-b) > epsilon:
    iterations += 1
    x_1 = a + 0.382 * (b - a)
    x_2 = a + 0.618 * (b - a)
    if y(x_1) <= y(x_2):
        b = x_2
    else:
        a = x_1
print('кол-во итераций', iterations)
print('result = ', y((x_1 + x_2) / 2))
```

Градиент - это вектор столбц частных производных вычисленных в точке.
Градиент направлен перпендикулярно
касательной плоскости, проведенной в точке х, в
сторону возрастания функции

```python
# зададим переменные и функцию
x1, x2 = symbols('x1 x2')

def function(x1,x2):
    return 2*x1**2 + x1*x2 + x2**2
y = function(x1,x2)
```

```python
# посиск производной 
def derivative_of_two_var():
    yprime = [y.diff(i) for i in [x1, x2]]
    return yprime

yprime = derivative_of_two_var()  
```

```python
# вычислим градиет в точке
def grad_in_point(point):
    nabla = lambdify((x1, x2), yprime, 'numpy')
    nabla_in_x = nabla(*point)
    return np.array(nabla_in_x)
```

```python
# вернет норму евклида
def euclidean_norma(nabla_in_point):
    return sum(map(lambda x: x**2 , nabla_in_point))**(1/2)
```

```python
def find_new_point(first, next_):
    return first  - h *next_
```

```python
# зададим шаг 
h  = 0.3
epsilon = 0.0001

# зададим начальную точку
x_1 =np.array([.5, 1])

iteration = 1 
status = True
point = x_1
while status:
    print('номер итерации: ', iteration)
    print('точка ', point)
    print(f'значение функции в этой точке {function(*point):.4f} ')
    # вычислим градиет в точке
    nabla_in_point = grad_in_point(point = point)
    # первое условие проверки остановки 
    if  euclidean_norma(nabla_in_point) > epsilon:
        new_point = find_new_point(point, nabla_in_point)
        if norm(new_point-point) > epsilon and np.abs(function(*new_point)-function(*point)) > epsilon:
            point = new_point
        else:
            status = False
            print('Solution was found')
    iteration +=1   
    print('-'*40)
        
```
