Меры изменчивости характеризуют различия между отдельными значениями выборки.

## Меры

### Размах
### $R = X_{max} - X_{min}$

Если есть длинный ряд чисел и его Xmax = 157, а Хmin = 185, то 

$R = 185-157 = 28$

Минус размаха в том, что он использует только два значения. Любое изменение очень болезненно скажется на размахе. Если мы добавим 3000 в конце, то размах из 28 станет 2843. Поэтому стоит использовать каждое значение.

### Дисперсия
Дисперсия - средний квадрат отклонений индивидуальных значений признака от их средней величины.

Дисперсия - один из способов использовать все значения для расчета. Она показывает на сколько в среднем данные отклоняются от среднего значения. 

Примерно так! [4, 5, 4, 5] 

Среднее 4,5.

Отклонения: -0.5, 0.5, -0.5, 0.5. 
> 4 - 4,5 = -0,5     
5 - 4,5 = 0,5 

Только для расчета дисперсии надо усреднить все отклонения. И тут проблема!

$-0.5 + 0.5 - 0.5 + 0.5 = 0$

$0 : n = 0$

Наше уровнение имеет вид $\Sigma^{n} _{i=1}(x_i - \overline{X})$

А как мы помним есть правило:
### $\Sigma^{n} _{i=1}(x_i - \overline{X}) = 0$  

Чтобы решить проблему надо избавиться от отрицательных значений, найдя квадрат от отклонений. Теперь выведем формулу десперсии. Нет нечего проще!

### $D = \frac{\Sigma^{n}_{i=1}(x_i-\overline{X})^2}{n-1}$ `-1 Здесь потому, что так точнее. Нам не объясили.`

### $D = \frac{\Sigma^{n}_{i=1}(x_i-M_i)^2}{n}$

У дисперсии есть одна проблема. Поскольку мы взяли квадрат, у нас сейчас тоже все в квадрате. Это нам надо решить следующей мерой.

### Среднее квадратичное отклонение

Тут слов не надо, хватит формул.

### $\sigma = \sqrt{D}$ `Тут D взят из выборки`

### $sd = \sqrt{D}$ `Тут D из генеральной совакупности.`

### Cреднее линейное отклонение

Это редко используется. но знать надо. Вместо квадрата для уничтожения отрицательных часто используют модуль:

$|-4| = 4$


$|-3500| = 3500$

## Если что-то не ясно

### Простое представление

#### $[1, 2, 2, 3, 4, 4, 5]$ 
#### $\overline{X} = 3$      
#### $n = 7$          
 $(1-3)^2 = 4$          
 $(2-3)^2 = 1$          
 $(2-3)^2 = 1$          
 $(3-3)^2 = 0$                                                    
 $(2-3)^2 = 1$          
 $(2-3)^2 = 1$          
 $(1-3)^2 = 4$     

### $D = \frac{4+4+1+1+1+1+0}{7-1} = \frac{12}{6} = 2$
### $sd = \sqrt{2} \approx 1.4$
