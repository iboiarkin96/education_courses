# Неделя 1

[Лекция](https://github.com/bdemeshev/coursera_metrics/tree/master/lec_01)

**Модель** - формула для объясняемой переменной

**RSS** – остаточная сумма квадратов (сумма квадратов остатков) – Residual Sum of Squares $\Sigma (y_i - \hat y_i) ^2= \Sigma \hat \varepsilon_i^2$ - проекция вектора $y$ на единичный вектор $ \hat y_i$

**TSS** - общая сумма квадратов Total Sum of Squares $\Sigma (y_i - \overline y) ^2$ -показывает насколько реальные значения перменной лежат далеко от среднего значения - проекция вектора $y$ на единичный вектор

**ESS** - объясненная сумма квадратов EXplained Sum of Squares  $\Sigma (\hat y_i - \overline y) ^2$ -показывает насколько прогназируемые значения перменной лежат далеко от среднего значения  - проекция вектора $ \hat y_i$ на единичный вектор

$\frac {ESS}{ TSS} = R^2= cos\phi^2 \in  [0, 1]$

TSS = RSS +  ESS по теореме пифагора 

$R^2$ - это
- доля объяснённого разброса в общем разбросе
- квадрат выборочного коэффициента корреляции
- $\frac{ESS}{TSS}$
- $cos^2(\phi)$

Регрессионая прямая проходит через геометрический центр (через $\hat x$ и  через $\hat y$), поэтому выбросы смещают оценки МНК, так как цент будет смещен относительного всего облака точек

Коэффициенты линейной регрессии с константой $y_i = \beta_1  + \beta_2*x_i  + \epsilon_i$
$$\hat\beta_2 = \frac{\Sigma(x_i - \overline{x})(y_i - \overline{y})}{\Sigma(x_i - \overline{x})^2}$$
$$\hat\beta_2 = \overline{y} - \hat\beta_2 * \overline{x}$$

```python
data = pd.read_csv('data//mtcars.csv', index_col=0)
data = data[['disp', 'wt']]

# чтобы посчитать регрессию найдем среднее значение y (disp) и x (wt)
disp_mean, wt_mean = data['disp'].mean(), data['wt'].mean()
data['y_i - y_avg'] = data['disp'] - disp_mean
data['x_i - x_avg'] = data['wt'] - wt_mean
# находим коэффициент бетта 
# находим числитель
numerator = sum(data['y_i - y_avg'] * data['x_i - x_avg'])
# находим знаменатель
denumerator  = sum((data['wt'] - wt_mean)**2)
```

```python
# рассчитывает бета2
wt_coeff = numerator/denumerator
wt_coeff
```

```python
# находим интерсепт 
Intercept = disp_mean-wt_coeff*wt_mean
Intercept
```

```python
# Регересисонная прямая Объем двигателя пути и веса проверка
# disp: Объем двигателя и  wt: 	вес
```

```python
import statsmodels.formula.api as smf

# линейная регрессия
model = smf.ols('disp ~ wt',data=data)
fitted = model.fit()

# коеффиценты
fitted.params

# выкладка по модели
print (fitted.summary())
```

```python
#остатки
data['resid']= fitted.resid
# предсказанные
data['pred'] = fitted.predict()
```

```python
RSS = sum((data['disp'] - data['pred'])**2)
TSS = sum((data['disp'] - data['disp'].mean())**2)
ESS= sum((data['pred'] - data['disp'].mean())**2)
```

```python
# коэффициент детерминации (R^2) через ESS и TSS
ESS/TSS
```

# Неделя 2

1) Ковариа́ция (корреляционный момент, ковариационный момент) — в теории вероятностей и математической статистике мера линейной зависимости двух случайных величин. 

$$cov (X, Y) = M [(X − M (X)) (Y − M (Y))]$$

2) Условное математическое ожидание $E(s | r) = \tilde s = f(r) $ - это такая функция от случайной величины r, которая наибольше похожа на s

$\tilde s $  очень похожа на $s$: $E(s) = E(\tilde s)$

$E(y_i | x_i)$ - это наилучший прогноз $y_i$, формулируемый с помощью $x_i$

**Статистические свойства оценок МНК**

$$E(\varepsilon_i | X) = 0$$
$$E(\varepsilon_i^2 | X) = \sigma^2 \;\; \; или\;\; \; \;    Var(\varepsilon_i | X) = \sigma^2$$
$$E(\varepsilon_i\varepsilon_j | X) = 0 \;\; \; или\;\; \; \;    Cov(\varepsilon_i\varepsilon_j  | X) = 0$$

Модель геометрической интерпретации
   - коварация: скалярное произведение случайных величин s и r
   - кореляция: косинус угла между случаными величиными

 При скалярном произведении s и r равном ковариации (то еcть $(s, r) = cov(s, r)$), перпендикулярность и равенство косинуса нулю, и некоррелированность --- это одно и то же.

Коварационная матрица вектора $\varepsilon$

$$Var(\varepsilon)=\begin{pmatrix}
  Var(\varepsilon_1) & Cov(\varepsilon_1, \varepsilon_2) & Cov(\varepsilon_1, \varepsilon_3)\\
  Cov(\varepsilon_2, \varepsilon_1)& Var(\varepsilon_2) & Cov(\varepsilon_2, \varepsilon_3)\\
  Cov(\varepsilon_3, \varepsilon_1)& Cov(\varepsilon_3, \varepsilon_2)& Var(\varepsilon_3)
\end{pmatrix}$$

Запись стат свойств МНК при помощи ковариациооной матрицы
$$Var(\varepsilon)=\begin{pmatrix}
  \sigma^2 & 0 & 0\\
  0 & \sigma^2  & 0\\
  0 & 0 & \sigma^2 
\end{pmatrix}= \sigma^2 \begin{pmatrix}
  1 & 0 & 0\\
  0 & 1  & 0\\
  0 & 0 & 1 
\end{pmatrix}$$

**Как оценить дисперсию и ковариацию коэффициентов регресии?** Для парной регресии 
$$Var(\hat\beta_1) = \sigma^2 *\frac{\Sigma{x_i^2}}{n *\Sigma(x_i - \overline{x})^2}$$
$$Var(\hat\beta_2) =  \frac{\sigma^2}{\Sigma(x_i - \overline{x})^2}$$


То есть другими словами $Var(\hat\beta_j | X) =  \sigma^2 * f(X)$

$$Cov(\hat\beta_1, \hat\beta_2) =  \sigma^2 * \frac{-\overline{x}}{\Sigma(x_i - \overline{x})^2}$$

**Оценка стандартного отклонения**- это $$\hat\sigma^2 = \frac{RSS}{n-k}$$, где n - кол-во наблюденений, k - колличство регрессоров 

тут есть зависимость чем больше объясняющих переменных, тем больше дисперсия


Имея оценку $\sigma^2$ - оценку стандартного отклонения можно получить оценку дисперсии коэффициента регресии 
$$\hat{Var(\hat\beta_j | X)} =  \hat{\sigma}^2 * f(X) = \frac{\hat{\sigma}^2}{RSS_j}$$
где $RSS_j$ - RSS  регресии j- ой объясняющей переменной на все остальные регрессоры

Среднее квадратическое отклонение (std err) - это корень из оценки дисперсии, показывает средние квадратические отклонения коэффициентов регрессии от их истинных значений $$se(\hat\beta_j) = \sqrt{\hat{Var(\hat\beta_j | X)}}$$

Чем меньше стандартная ошибка тем точнее оценка.

Рассчитаем стандартную ошибку коэфиицентов дисперсии 

```python
# оценка для бета_2
import math
sigma_sqrt = RSS /(data.shape[0] - 1 -1 )
# (n - m - 1), где n – число наблюдений, m – число независимых переменных.
denumerator_of_se = sum((data.wt - data.reset_index().wt.mean())**2)
se = math.sqrt(sigma_sqrt / denumerator_of_se)
se
```

```python
# оценка для бета_1
numerator_of_se = sum((data.wt)**2)
se = math.sqrt(sigma_sqrt * numerator_of_se / (denumerator_of_se * data.shape[0]))
se
```

# Список хороших свойст МНК

 - *базовые (верные без предположения о нормальности )*
  -  Оценки $\hat\beta_j$ линейны по $y_i$
  - Оценки не смещены $E(\hat\beta_j|X) = \beta_j$ и $E(\hat\beta_j) = \beta_j$
  - Оценки эффектины среди линейных и несмещенных  (любая другая оцнка будет имень большую дисперсию чем оценки полученнымые при помощи МНК)
 - *ассимптотические (верны на больших выборках даже без предположения о нормальности ошибок)*
  - оценки $\beta_j$ при  n -> бесконечности стремятся по вероятности к истинным оценкам 
       $$\frac{\hat\beta_j - \beta_j}{se(\hat\beta_j)}) -> N(0, 1)$$
       $$\hat\sigma^2->\sigma^2$$
 - *при нормальности (верны на маленьких выборках при нормальности ошибок)*
   - оценки $\beta_j$ эффективны среди всех несмещенных
   - имеет распределенние Стьюдента
       $$\frac{\hat\beta_j - \beta_j}{se(\hat\beta_j)}) \sim t_{n-k}$$
   - имеет распределенние Хи-квадрат 
       $$\frac{RSS}{\sigma^2} \sim \chi^2_{n-k}$$
       

# Предпосылки МНК
 - истинная зависимость имеет вид, который мы оцениваем (линейная, квадратичная)
 - регрессия стоится на те регрессоры, от которых дейсвительно зависит объясняемая переменная
 - наблюденйи больше чем регрессоров
 - ограничения на ошибки 
     - строгая экзогенноть $E(\varepsilon_i | X)$ = 0
     - условная гомоскедастичность $Var(\varepsilon_i | X) = \sigma^2$
     - отсутвие условной корреляции $Cov(\varepsilon_i\varepsilon_j|X) = 0$
 - ограничения на регрессоры
      - отсутвие мультиколениарности (нет роста в см и в метрах одновремено). Это можно задасть как $\det(X'X) != 0$
      - векторы отдельных наблдеений - независимы и одинаково распредлены
