# Что такое ноутбуки и как ими пользоваться

```python
# факториал реализация 1
def f1(x):
    y = 1
    for i in range(1, x+1):
        y*=i
    return y
```

```python
# факториал реализация 2
def f2(n):
    if n <= 1:
        return 1
    else:
        return (n * f2(n-1))
```
