# Что такое ноутбуки и как ими пользоваться

```python
# факториал реализация 1
def f1(x):
    y = 1
    for i in range(1, x+1):
        y*=i
    return y
```

```python
# факториал реализация 2
def f2(n):
    if n <= 1:
        return 1
    else:
        return (n * f2(n-1))
```

```python
# built in 
from math import factorial
factorial(5)
```

```python
import timeit

boz = []
for i in range(1, 100, 1):
    try:
        boz.append([i,
                    float(timeit.timeit("f1(i)", setup="from __main__ import f1, i", number=10000)),
                    float(timeit.timeit("f2(i)", setup="from __main__ import f2, i", number=10000)),
                    float(timeit.timeit("factorial(i)", setup="from __main__ import factorial, i", number=10000))])
    except:
        pass


import pandas as pd
pd.DataFrame(boz).rename(columns = {0: 'count',
                                    1: 'цикл',
                                    2: 'рекурсия',
                                    3: 'built-in'}).plot(x = 'count')
```
