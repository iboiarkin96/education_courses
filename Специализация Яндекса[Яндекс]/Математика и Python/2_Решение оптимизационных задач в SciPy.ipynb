# Решение оптимизационных задач в SciPy (Версия для Python 3)

```python
from scipy import optimize
```

```python
def f(x):   # The rosenbrock function
    return .5*(1 - x[0])**2 + (x[1] - x[0]**2)**2
    
print(f([1, 1]))
```

```python
# простой пербор 
result = optimize.brute(f, ((-5, 5), (-5, 5)))
print(result)
```

```python
# дифференциальная эволюция - ген алгоритм
print(optimize.differential_evolution(f, ((-5, 5), (-5, 5))))
```

```python
# если у функции есть градиент
# то можно воспользоваться градиентными методами

# Градиент функции - это вектор координатами которого являются частные производные этой функции по всем её переменным.
```

```python
import numpy as np
# функция разенброка
def g(x):
        return np.array((-2*.5*(1 - x[0]) - 4*x[0]*(x[1] - x[0]**2), 2*(x[1] - x[0]**2)))
```

```python
# сравним вычислительно градиент функйци и расчитанный градиент 
# видим что расхождение минимально
print(optimize.check_grad(f, g, [2, 2]))
```

```python
# метод bfgs

# Относится к классу так называемых квазиньютоновских методов. В отличие от ньютоновских методов в квазиньютоновских не вычисляется напрямую гессиан функции, т.е. нет необходимости находить частные производные второго порядка. Вместо этого гессиан вычисляется приближенно, исходя из сделанных до этого шагов.
print(optimize.fmin_bfgs(f, [2, 2], fprime=g))
```

```python
# можно не задавать метод оптимизации
print(optimize.minimize(f, [2, 2]))
```

```python
# Как c помощью модуля optimize библиотеки scipy можно найти максимум функции f(x)? 

def g(x):
  return -f(x)

optimize.minimize(g, [1, 1])
```

```python
print(optimize.minimize(f, [2, 2], method='BFGS'))
```

```python
print(optimize.minimize(f, [2, 2], method='Nelder-Mead'))
```
